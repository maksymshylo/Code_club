## Перевірка банківської карти

**Алгоритм Луна** (англ. Luhn algorithm), або формула Луна (англ. Luhn formula), також відомий під назвою **«modulus 10»** або «mod 10», це проста формула перевірки контрольної суми, що використовується для валідації різноманітних ідентифікаційних номерів, таких як

- номери кредитних/платіжних карток, 
- номери IMEI,
- американських National Provider Identifier Number, 
- канадських Canadian Social Insurance Number,
- ізраїльських ID Numbers та грецьких Social Security Numbers (ΑΜΚΑ).

Алгоритм був створений науковцем з IBM **Гансом Петером Луном** і описаний патентом U.S. Patent No. 2,950,048, описаним 6 січня 1954 року та схваленим 23 серпня 1960 року.

Алгоритм є публічним і широко використовується. Він також вказаний у ISO/IEC 7812.

Цей алгоритм не створювався як криптографічно надійна хеш-функція, а суто як захист від випадкових помилок.

Більшість кредитних карток і багато урядових ідентифікаційних номерів використовують цей алгоритм як простий метод відсіювання неправильних номерів.
    

### Алгоритм
______
Формула перевіряє номер за його контрольною цифрою, котра зазвичай додається до часткового номера акаунту (partial
account number) при генеруванні повного номера акаунту (full account number). Цей номер повинен успішно проходити таку
перевірку:
    
1. Починаючи від крайньої правої цифри (контрольної), рухаємося ліворуч, подвоюючи кожну другу цифру. 
Якщо результат подвоєння є більшим, ніж 9 (як-от: 8 × 2 = 16), то потрібно додати числа результату (наприклад: 16:
1 + 6 = 7, 18: 1 + 8 = 9), або, як альтернатива — відняти 9 від результату (16: 16 − 9 = 7, 18: 18 − 9 = 9).
2. Знайти суму всіх цифр.
3. Якщо ділення суми на 10 не має остачі (сума закінчується нулем,) то номер є правильним відповідно до
формули Луна.

#### Приклад_1
```
4  5  6  1     2  6  1  2     1  2  3  4     5  4  6  4  - початковий номер
8     12       4     2        2     6        10    12    - вибрано кожна друга цифра з кінця і помножена на 2
8     3        4     2        2     6        1     3     - якщо ця цифра > 9 змінюємо: цифра = цифра - 9
```
Потім всі числа додаються:
```
8+ 5+ 3 +1  +  4+ 6+ 2+ 2  +  2+ 2+ 6+ 4  +  1+ 4+ 3+ 4 = 57
```
Отримана сума має ділитись націло(без остачі) на 10

У даному випадку 57 не ділиться без остачі на 10

Значить номер **не** пройшов перевірку
______

#### Приклад_2
```
4  5  6  1     2  6  1  2     1  2  3  4     5  4  6  7  - початковий номер
8     12       4     2        2     6        10    12    - вибрано кожна друга цифра з кінця і помножена на 2
8     3        4     2        2     6        1     3     - якщо ця цифра > 9 змінюємо: цифра = цифра - 9
```
Потім всі числа додаються:
```
8+ 5+ 3 +1  +  4+ 6+ 2+ 2  +  2+ 2+ 6+ 4  +  1+ 4+ 3+ 7 = 60
```
Отримана сума має ділитись націло(без остачі) на 10

У даному випадку 60 ділиться без остачі на 10

Значить номер **пройшов** перевірку

````python
#Алгоритм Луна
def checking_card(number):
    number.reverse()                     # обертає список number
    x = []
    y = []
    for i in range(1,len(number), 2):    # вибирає кожне друге
        x.append(number[i])              # записує в список x 
    for i in range(0,len(x)):    
        x[i] = 2*x[i]
        if x[i] > 9:
            x[i] = x[i] - 9
    x = sum(x)                           # вбудована функція sum() сумує всі елементи списку
    for i in range(0,len(number),2):     # вибирає кожне третє
        y.append(number[i])              # записує в список y 
    y = sum(y)
    print("Сума", x + y)
    if (x + y) % 10 == 0:
        return True
    else:
        return False
````
