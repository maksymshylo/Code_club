### Цикли - це інструкції, які виконують одну й ту ж саму послідовність дій, поки діє задана умова.

У реальному житті ми досить часто стикаємося з циклами.

Наприклад, ходьба людини - цілком циклічне явище: крок лівою, крок правою, знову лівою-правою і т.д., поки не буде досягнута певна мета (наприклад, школа або магазин).

У комп'ютерних програмах поряд з інструкціями багатьох виборів також існують інструкції циклів (повторення дії).

Якби інструкцій циклу не існувало, довелося б багато разів вставляти в програму один і той же код поспіль стільки раз, скільки потрібно виконати однакову послідовність дій.

### Універсальним організатором циклу в мові програмування Python (як і в багатьох інших мовах) є конструкція while. Слово "while" з англійської мови перекладається як "поки" ( "поки логічне вираження повертає істину, виконувати певні операції").

Інструкція while є найуніверсальнішою конструкцією організації ітерацій в мові Python. Простіше кажучи, вона продовжує виконувати блок інструкцій (зазвичай з відступами) до тих пір, поки умовний вираз продовжує повертати істину. Вона називається «циклом», тому що управління циклічно повертається до початку інструкції, поки умовний вираз не поверне помилкове значення.

Як тільки в результаті перевірки буде отримано помилкове значення, управління буде передано першої інструкції, розташованої відразу ж за вкладеним блоком тіла циклу while. В результаті тіло циклу залишиться активним знову і знову, поки умовний вираз повертає

```python
while умова:
    блок інструкцій
```

#### При виконанні циклу while спочатку перевіряється умова.

#### Якщо вона помилкова(False), то виконання циклу припиняється і керування передається на наступну інструкцію після тіла циклу while.

#### Якщо умова істинна(True), то виконується інструкція, після чого умова перевіряється знову і знову виконується інструкція.

#### Так триває до тих пір, поки умова буде істинною. Як тільки умова стане помилково, робота циклу завершиться і управління передасться наступної інструкції після циклу.

Для ілюстрації розглянемо декілька простих циклів **while** в дії. Перший, який містить інструкцію **print**, вкладену в цикл while, просто виводить повідомлення до нескінченності. 

Не забувайте, що **True** - це всього лише особлива версія цілого числа **1**, і воно означає справжнє значення, тому результатом цього умовного виразу завжди буде істина, і інтерпретатор нескінченно буде виконувати тіло циклу, поки ви не скасуєте його виконання. Такі цикли зазвичай називаються **нескінченними**:

```python
while True:
    print('Type Ctrl-C to stop me!') # рекомендую запускати в сmd, бо саме там працює команда Ctrl-C 
```

Наступний фрагмент продовжує вирізати з рядка перший символ, поки вона не спорожніє і в результаті не перетвориться в помилкове значення.

Це звичайна практика - перевірка істинності об'єктів здійснюється безпосередньо замість використання більш розтягнутого еквівалента **(while x! = '' :)**.

```python
x = 'spam'
while x:    # поки x не пустий рядок
    print(x, end = ' ')
    x = x[1:]
```

Наступний фрагмент перебирає значення від **a до b**, не включаючи значення **b**.

У наступному уроці ми розглянемо більш простий спосіб виконання цих же дій за допомогою циклу **for** і вбудованої функції **range**:

```python
a = 0
b = 10
while a < b:
    print(a)
    a+=1                      #теж саме що і: a = a +1
```

```python
import time
s = "+" 
i = 0 
while i < 10:
    print (s)
    time.sleep(1)
```

В останньому рядку коду відбувається збільшення значення змінної i на одиницю, тому з кожним оборотом циклу її значення збільшується. 

Коли буде досягнуто число 10, логічне вираз i <10 дасть помилковий результат, виконання тіла циклу буде припинено, а потік виконання програми перейде на команди наступні за всією конструкцією циклу.

Результатом виконання скрипта наведеного вище є висновок на екран десяти знаків + в стовпчик. Якщо збільшувати лічильник в тілі циклу не на одиницю, а на 2, то буде виведено лише п'ять знаків, так як цикл зробить лише п'ять оборотів.

```python
x = 1
while x <= 10:
    print(x ** 2)
    x = x + 1
```

```python
# Кількість  цифр  натурального числа
n = int(input())
length = 0
while n > 0:
    n //= 10  # це єквівалентно n = n // 10
    print(n)
    length += 1
print(length)
```

Після тіла циклу можна написати слово else: і після нього блок операцій, який буде виконаний один раз після закінчення циклу, коли перевіряється умова стане невірно:

```python
i = 1
while i <= 10:
    print(i)
    i += 1
    time.sleep(1)
else:
    print('Цикл закінчений, i =', i) # як ви бачите функція виводу print() може виводити більше ніж один аргумент
```

## Множинне присвоювання

У Пітоні можна за одну інструкцію присвоювання змінювати значення відразу декількох змінних. Робиться це так:

```python
a, b = 0, 1
```

```python
a
```

```python
b
```

```python
a = 0
b = 1
```

Відмінність двох способів полягає в тому, що множинне присвоювання в першому способі змінює значення двох змінних одночасно.
Якщо зліва від знака «=» в множині присвоєнні повинні стояти через кому імена змінних, то справа може стояти довільні вирази, розділені комами. Головне, щоб зліва і праворуч від знака присвоювання було однакове число елементів.

Множинне присвоювання зручно використовувати, коли потрібно обміняти значення двох змінних. У звичайних мовах програмування без використання спеціальних функцій це робиться так:

```python
a = 1
b = 2
tmp = a
a = b
b= tmp
print(a, b)
```

У Пітоні тією ж дію записується в один рядок:

```python
a = 1
b = 2
a, b = b, a
print(a, b) 
```
